**IMPORTO LIBRERIAS**

```python

import ast              # module Abstract Syntax Trees
import json             # module json encoder and decoder
import pandas as pd     # module panda Data Analysis Library
import nltk             # module nlkt - natural language toolkit
```

**CARGO ARCHIVO JSON A UN DATASETS**

```python
# list 'output_steam_games.json'
rows = []
with open ('..\\JSON\\output_steam_games.json', 'r', encoding='utf-8') as f:
    for line in f.readlines():
        rows.append(json.loads(line))

# save to dF
df = pd.DataFrame(rows)
# tiempo 2.3s
```

**ELIMINAMOS LOS REGISTROS CON VALORES NAN Y DUPLICADOS**

```python
#Eliminamos cualquier registro que tenga al menos un NaN
df.dropna(how='any', inplace=True)
```

```python
valores_nulos_por_columna = df.isnull().sum()
valores_nulos_por_columna
```

```python
valores_nulos_por_columna = df.isnull().sum()
valores_nulos_por_columna
```

**ELIMINO COLUMNAS QUE NO VAN A SER USADAS**

```python
# elimino las columnas que no  necesito
df = df.drop(['publisher', 'app_name', 'url', 'tags', 'reviews_url', 'specs', 'early_access'], axis=1)
```

**ANALIZAMOS LOS DATOS DE LAS COLUMNAS**


 ---- *Columna GENRES*

```python
# Vemos como esta compuesta la columna genres
df['genres'].value_counts
```

```python
# Desanidamos la lista dentro de la columna
df = df.explode('genres')
```

```python
# Pasamos a variables dummies el contenido de genres
dummies_genres = pd.get_dummies(df['genres'], prefix='genres')
```

```python
# Concatenar las variables dummies al DataFrame original y eliminar la columna original
df = pd.concat([df, dummies_genres], axis=1)
df.drop('genres', axis=1, inplace=True)
```

```python
# Verificamos el cambio. 
df.head(3)
```

```python
# Tomamos a partir de la quinta columna y cambiamos los True o False a 1 y 0
df[df.columns[5:]] = df[df.columns[5:]].astype(int)
```

 ---- *Columna RELEASE_DATE*

```python
# se extrae el año unicamente de la columna release date
df['release_date'] = df['release_date'].str.extract(r'(\d{4})')
```

```python
df.head(2)
```

 ---- *Columna PRICE*

```python
df['price'].unique()
```

```python
# paso todos los valores a string
df['price'] = df['price'].astype(str)
```

```python
# Cambio todos los string que contienen Free por 0.00
condicion = df['price'].str.contains('Free')
df.loc[condicion, 'price']='0.00'
```

```python
# Cambio todos los string que contienen Install por 0.00
condicion = df['price'].str.contains('Install')
df.loc[condicion, 'price']='0.00'
```

```python
# Cambio todos los string que contienen party por 0.00
condicion = df['price'].str.contains('party')
df.loc[condicion, 'price']='0.00'
```

```python
# Cambio todos los string que contienen Play por 0.00
condicion = df['price'].str.contains('Play')
df.loc[condicion, 'price']='0.00'
```

```python
# paso todos los valores de string a float
df['price'] = df['price'].astype(float)
```
